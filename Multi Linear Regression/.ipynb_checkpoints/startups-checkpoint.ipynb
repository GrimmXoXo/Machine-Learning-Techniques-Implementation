```python
def iqr(x):
        q25=x.quantile(0.25)
        q75=x.quantile(0.75)
        iq=q75-q25
        lower=q25-(1.5*iq)
        upper=q75+(1.5*iq)
        return x[(x>upper)|(x<lower)]
```

## 85% accurate model

```python
#Now lets try to use Gradient Descent To calculate the prediction X_test
#As this has multiple coeff_ or weights we will try to get a generalized equation for the whole dL/d(features) instead of 
#Calculating for each and every single feature seperately
#Parameters-> m,b
#Hyperparameters-> epochs,learning rate



class GDRegressor:
    def __init__(self,epochs=100,learning_rate=0.01):
        self.epochs=epochs
        self.learning_rate=learning_rate
        self.coef_=None
        self.intercept_=None
    
    def fit(self,X,y):
        self.coef_=np.ones((X.shape[1],1))#The number of columns will be equal to number of feature weights
        self.intercept_=0 #Initializing weight of intercept as 0
        for i in range(self.epochs):
            yhat=np.dot(X,self.coef_)+self.intercept_

            #Lets focus on intercept first
            dLb0=-2*np.mean(y-yhat)
            self.intercept_=self.intercept_-(self.learning_rate*dLb0)

            #Now lets focus on coeff
            dLdb=-2*np.dot((y-yhat).T,X).T/X.shape[0]
            self.coef_=self.coef_-(self.learning_rate*dLdb)

    
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_)+self.intercept_
```

```python
gd=GDRegressor(epochs=100,learning_rate=0.1)
gd.fit(X_train,y_train)
```

```python
r2_score(y_test,gd.predict(X_test))
```
